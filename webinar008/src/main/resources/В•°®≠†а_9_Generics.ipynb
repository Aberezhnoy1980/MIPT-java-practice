## 1. Определения



Обобщения (Generics) позволяют указать ограничение, накладываемое на поведение класса или методов, в терминах неизвестных типов

### Демонстрация работы «*Применение обобщения*»

<br>

Напишем код:

```java
// Нора
class Burrow {

    // Житель
    Object habitant;

    public void setHabitant(Object habitant) {
        this.habitant = habitant;
    }

    public Object getHabitant() {
        return this.habitant;
    }
}

class Rabbit {
}

class Main {
    public static void main(String[] args) {
        Burrow burrow = new Burrow();
        // указываем жителя
        burrow.setHabitant(new Rabbit());
        // Нужно явное приведение типа!
        Rabbit rabbit = (Rabbit) burrow.getHabitant();
    }   
}
```

Так как класс работает с `Object`, вы можете свободно поместить в нору <p>`Burrow` абсолютно любой класс. Нет никакого способа проверить использование <p>класса во время компиляции.

## 2. Соглашение об именовании





По соглашению переменные типа именуются одной буквой в верхнем регистре.

`E` — Элемент

`K` — Ключ

`N` — Число

`T` — Тип

`V` — Значение

`S`, `U`, `V` и т. п.

## 3. Бриллиантовая операция



Начиная с Java 7 существует бриллиантовая операция (diamond operator), которая позволяет указывать пустые аргументы типа <> там, где компилятор может вывести тип из контекста:

```java
Burrow<Rabbit> burrow = new Burrow<>();
```

## 4. Обобщённые методы



Обобщённые методы похожи на обобщённые классы, но параметры типа относятся к методу, а не к классу. Допустимо делать обобщёнными статические и нестатические методы, а также конструкторы:

```java
class Utils {
    // Обобщённый метод.
    static <T> void setIfNull(Burrow<T> burrow, T t) {
        if (burrow.getHabitant() == null) {
            burrow.setHabitant(t);
        }
    }
}
```

```java
Burrow<Rabbit> burrow = new Burrow<>();
Utils.setIfNull(burrow, new Rabbit());
```

## 5. Ограничение параметра типа



В некоторых случаях имеет смысл ограничить  типы, которые можно использовать в качестве аргументов в параметризованных типах. Допустим, в норе могут жить только наследники класса `Animal`. Подобное ограничение можно сделать с помощью ограниченного параметра типа (bounded type parameters).

Чтобы объявить ограниченный параметр типа, нужно после имени параметра указать ключевое слово `extends`, а затем указать верхнюю границу (upper bound), которой в данном примере является класс `Animal`.

В этом контексте `extends`  означает как `extends`, так и `implements`.

### Демонстрация работы «*Ограничение*»

<br>

Напишем код:

```java
class Burrow<T extends Animal> {

    T habitant;
    
    public void setHabitant(T habitant) {
        this.habitant = habitant;
    }

    public T getHabitant() {
        return this.habitant;
    }

    public void tick() {
        if (habitant != null) {
            // Можно вызывать методы
            // интерфейса или класса,
            // указанного в качестве верхней
            // границы параметра типа.
            habitant.doSomething();
        }
    }

}

class Animal {
    public void doSomething() {
        System.out.println("Doing something.");
    }
}

class Rabbit extends Animal {
}

class Main {
    public static void main(String[] args) {
        Burrow<Rabbit> burrow = new Burrow<>();
        burrow.setHabitant(new Rabbit());
        burrow.tick();
    }
}

```

Попробуем создать класс, не являющийся наследником `Animal`, и вызвать <p>`setHabitant`.

## 6. Наследование и обобщение

### Обобщения, наследование и дочерние типы

Как вы уже знаете, можно присвоить объекту одного типа объект другого типа, если эти типы совместимы. Например, вы можете присвоить объект типа `Integer`  переменной типа `Number`, так как `Number` является одним из супертипов `Integer`.

```java
public void someMethod(Number n) {
  /* ... */
}

someMethod(new Integer(10));   // OK
someMethod(new Double(10.1));   // OK
```

Это также верно для обобщений. Вы можете осуществить вызов обобщённого типа, передав `Number` в качестве аргумента типа, и любой дальнейший вызов будет разрешён, если аргумент совместим с `Number`:

```java
List<Number> list = new List<Number>();
list.add(new Integer(10));   // OK
list.add(new Double(10.1));  // OK
```

Вы можете указать обобщённый класс или интерфейс в качестве родительского для своего класса или интерфейса.

Так, например, для коллекций:

- `ArrayList<E> implements List<E>`,

Поэтому `ArrayList<String>` является дочерним для `List<String>`, который является дочерним для `Collection<String>`


### Подстановочный символ



Несмотря на то, что `Integer` является подтипом `Number`, `List<Integer>` не является подтипом `List<Number>`. Это разные типы. Общим предком для `List<Number>` и `List<Integer>` является `List<?>`.

Чтобы создать такую связь между этими классами, чтобы код мог иметь доступ к методам `Number` через элементы `List<Integer>`, используйте подстановочный символ:

```java
List<? extends Integer> intList = new ArrayList<>();
List<? extends Number> numList = intList;
```

В обобщённом коде знак вопроса (?), называемый подстановочным символом, означает  неизвестный тип

#### Ограничение сверху

Чтобы объявить ограниченный сверху подстановочный символ, используйте символ вопроса «?» с последующим ключевым словом `extends` и последующим ограничением сверху.

Чтобы написать метод, который работает со списками `Number` и дочерними типами от `Number`, например `Integer`, `Double` и `Float`, вы можете указать `List<? extends Number>`.

`List<Number>` вводит более жёсткое ограничение, чем `List<? extends Number>`, потому что оно соответствует только списку типа `Number`, а  `List<? extends Number>` соответствует списку типа `Number` и спискам всех его подклассов.

#### Неограниченный подстановочный символ

Если просто использовать подстановочный символ, получится подстановочный символ без ограничений. `List<?>` означает список неизвестного типа.

Неограниченный подстановочный символ полезен в двух случаях:

- Вы пишете метод, который может быть реализован с помощью функциональности класса `Object`
- Код использует методы обобщённого класса, которые не зависят от параметра типа, например `List.size()` или `List.clear()`

#### Ограничение снизу

Ограниченный снизу подстановочный символ ограничивает неизвестный тип так, чтобы он был либо указанным типом, либо одним из его предков:

```java
public static void addNumbers(List<? super Integer> list) {
    for (int i = 1; i <= 10; i++) {
        list.add(i);
    }
}
```

## 7. java.util.List




Интерфейс `java.util.List` представляет собой упорядоченную коллекцию. `List` может содержать дублирующиеся элементы.

Java содержит два класса, реализующих интерфейс `List`:
- `java.util.ArrayList`
- `java.util.LinkedList`

Операции для работы с элементами по индексу:

- `E get(int index)` — возвращает элемент по индексу
- `E set(int index, E element)` — заменяет элемент по указанному индексу. Возвращает старый элемент
- `void add(int index, E element)` — вставляет элемент в позицию по указанному индексу
- `E remove(int index)` — удаляет элемент по указанному индексу
- `boolean addAll(int index, Collection<? extends E> c)` — вставляет все элементы из коллекции c в позицию по указанному индексу
- `int indexOf(Object o)` — возвращает индекс элемента `o`, либо `-1`, если нет элемента с таким индексом
