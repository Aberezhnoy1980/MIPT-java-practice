## 1. Условные операторы


### Операторы if-then и if-then-else

Позволяют выполнять оператор или блок операторов только при соблюдении определённого условия.

Имеют две формы:
* `if-then` позволяет выполнить кусок кода при соблюдении условия

```java
Scanner sc = new Scanner(System.in);

boolean isCracked = sc.nextBoolean();

if (isCracked)
    System.out.println("Cracked");

if (isCracked) {
    System.out.println("In block");
    System.out.println("Cracked");
}
```


* `if-then-else` позволяет в дополнение указать кусок, который выполнится, если условие не соблюдено

```java
Scanner sc = new Scanner(System.in);

boolean isCracked = sc.nextBoolean();

if (isCracked)
    System.out.println("Cracked");
else
    System.out.println("Not cracked");

if (isCracked) {
    System.out.println("In block");
    System.out.println("Cracked");
}
else {
    System.out.println("Not block");
    System.out.println("Not cracked");
}
```

В блоке `else` можно проверить другое условие:

```java
Scanner sc = new Scanner(System.in);

int x = sc.nextInt();

if (x < 0) {
    System.out.println("< 0");
} else if (x == 3) {
    System.out.println("== 3");
} else if (x > 10) {
    System.out.println("> 10");
} else {
    System.out.println("!!!");
}
```

Для проверки можно использовать абсолютно любое выражение, возвращающее `boolean`.

### Оператор switch

Рассмотрим код:

```java
Scanner sc = new Scanner(System.in);
int x = sc.nextInt();

if (x == 0) {
    // Инструкции для x == 0
} else if (x == 1) {
    // Инструкции для x == 1
} else if (x == 2) {
    // Инструкции для x == 2
}
else {
    // Инструкции для всех остальных случаев
}
```

Тот же пример, но с оператором `switch`:

```java
Scanner sc = new Scanner(System.in);
int x = sc.nextInt();

switch (x){
    case 0:
        // Инструкции для x == 0
        break;
    case 1:
        // Инструкции для x == 1
        break;
    case 2:
        // Инструкции для x == 2
        break;
    default:
        // Инструкции для всех остальных случаев
        break;
}
```

С помощью `if` можно проверять любое условие, а с помощью `switch` — только значения выражения типов `byte`, `short`, `char`, `int`, `enum`, `String`, `java.lang.Byte`, `java.lang.Short`, `java.lang.Character`, `java.lang.Integer`.

Проверяемые значения в `case` должны быть константными.

Ключевое слово `break` не обязательно. Если его нет, по завершении выполнения блока операторов внутри одного `case` выполнятся операторы следующего. Это позволяет использовать один блок операторов для нескольких значений:

```java
switch (x) {
    case 0:
        // Инструкции для x == 0
        break;
    case 1:
    case 2:
        // Инструкции для x == 1
        // Инструкции для x == 2
        break;
    default:
        // Инструкции для всех остальных случаев
        break;
}
```

В Java 14 появился switch expression, который позволяет записывать оператор более кратким способом, а также возвращать значения.

```java
Scanner sc = new Scanner(System.in);
int type = sc.nextInt();

String moneyType;
switch (type) {
    case 1:
    case 2:
        moneyType = "Gold";
        break;
    case 3:
        moneyType = "Dollar";
        break;
    default:
        moneyType = "Rubles";
        break;
}
```

Аналогичный ему код:

```java
Scanner sc = new Scanner(System.in);
int type = sc.nextInt();

String moneyType;
switch (type) {
    case 1, 2 -> moneyType = "Gold";
    case 3 ->  moneyType = "Dollar";
    default -> moneyType = "Rubles";
}
```

Для возвращения значения из блока используется ключевое слово `yield`:

```java
Scanner sc = new Scanner(System.in);
int type = sc.nextInt();

String moneyType = switch (type) {
    case 1, 2:  yield "Gold";
    case 3: yield  "Dollar";
    default: yield "Rubles";
};
```

### Тернарная операция

Операция `?:` назвается тернарной, т. к. принимает 3 операнда, в то время как унарная — 1, а бинарная — 2.

```java
Scanner cs = new Scanner(System.in);
int x = cs.nextInt();
        
String result = x > 10 ? "yes" : "no";
```

эквивалентна

```java
Scanner cs = new Scanner(System.in);
int x = cs.nextInt();

String result;
if (x > 10)
    result = "yes";
else
    result = "no";
```

## 2. Циклы









Циклы позволяют выполнять оператор или блок операторов несколько раз.

### Оператор while

Вычисляет условие и выполняет оператор или блок операторов, если условие равно `true`. Затем ещё раз вычисляет условие и проверяет его. Если оно равно `true`, то снова выполняет оператор / блок операторов.

Так происходит, пока условие не станет `false`.

```java
int i = 10;
while (i < 10) {
    System.out.println(i);
    i++;
}
```

### Оператор do-while

Сначала выполняет операторы, а потом проверяет условие, т. е. оператор / блок операторов выполнится как минимум один раз.

```java
int i = 0;
do {
    System.out.println(i);
    i++;
} while (i < 0);
```

### Оператор for

```java
for ([инициализация]; [условие]; [инкремент]) {
  [операторы]
}
```

* Инициализация выполняется только один раз перед началом итераций

* Условие выполняется перед каждой итерацией. Если `false`, то выполнение цикла завершается

* Операторы выполняются только в тех случаях, если условие `true`

* Инкремент выполняется после каждой итерации и перед проверкой условия


Любой из этих компонентов можно убрать:

```java
for (; ; ) ;
```

Вывод чисел:

```java
for (int i = 0; i < 10 ; i++) {
    System.out.println(i);
}
```

Можно указывать несколько значений через запятую:

```java
for (int n = 0, m = 3; n <= 10 ; n++, m--) {
    System.out.println("n = " + n + "; m = " + m);
}
```

## 3. Операторы перехода

Позволяют передавать управление в другую часть вашей программы.

### Оператор break

Позволяет прервать выполнение `switch` или циклов и перейти к следующему оператору после них.

```java
int n1 = 0;
while (true) {
    System.out.println(n1);
    if (n1 == 10)
        break;
    n1++;
}

int n2 = 0;
do {
    System.out.println(n2);
    if (n2 == 10)
        break;
    n2++;
} while (true);

for (int n3 = 0; ; n3++) {
    System.out.println(n3);
    if (n3 == 10)
        break;
}
```

В случае вложенных циклов оператор `break` прерывает выполнение самого глубокого из текущих. Можно прервать внешний цикл, если указать на него метку:

```java
outer_for:
for (int n = 0; n < 10; n++) {
    inner_for:
    for (int m = 0; n < 10; m++) {
        if (n == 9 && m == 4)
            break outer_for;
        
        if (n == 7 && m == 2)
            break;
        
        if (n == 9 && m % 2 == 1)
            break inner_for;
    }
}
```

### Оператор continue

Позволяет пропустить текущую итерацию и перейти к следующей в цикле. Работает с меткой или без неё.

```java
for (int n = 0; n < 10; n++) {
    if (n % 2 == 0) {
        continue;
    }
    System.out.println(n);
}
```

### Оператор return

Возвращает управление из метода в тот, который его вызвал.

Имеет две формы: с возвращаемым значением и без него (для методов с `void`).
